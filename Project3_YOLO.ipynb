```python
# Step 1: Install required libraries
!pip install ultralytics

# Step 2: Import necessary packages
from ultralytics import YOLO
from google.colab import drive
import os

drive.mount('/content/drive')
```

```python
# Step 4: Configure and load the YOLOv8 nano model
# Initialize the model with the YOLOv8-nano configuration
model = YOLO('yolov8n.pt')  # Ensure this model file exists or use a pre-trained version

# Step 5: Define the dataset location
DATASET_PATH = '/content/drive/MyDrive/AER850/data/data.yaml'  # Path to your dataset configuration file

# Step 6: Configure hyperparameters for training
EPOCHS = 185
BATCH_SIZE = 8
IMG_SIZE = 960
MODEL_NAME = "pcb_detection_yolov8n"  # Name for the trained model


```

```python
# Step 7: Train the model
model.train(
    data=DATASET_PATH,   # Path to the dataset configuration file
    epochs=EPOCHS,       # Number of training epochs
    batch=BATCH_SIZE,    # Batch size for training
    imgsz=IMG_SIZE,      # Input image size
    name=MODEL_NAME      # Model name for saving results
)
```

```python
import shutil
import os

# Define the model name
MODEL_NAME = "pcb_detection_yolov8n"  # Name for the trained model

# Define source and destination paths
source_folder = f'/content/runs/detect/{MODEL_NAME}/'
new_destination_folder = '/content/drive/MyDrive/AER850/data/evaluation'  # Your specified destination folder

# Ensure the new destination folder exists
os.makedirs(new_destination_folder, exist_ok=True)

# List of files to copy
files_to_copy = [
    'confusion_matrix.png',
    'confusion_matrix_normalized.png',
    'P_curve.png',
    'R_curve.png',
    'PR_curve.png',
    'F1_curve.png'
]

# Copy each file to the new destination folder
for file_name in files_to_copy:
    shutil.copy(os.path.join(source_folder, file_name), new_destination_folder)
```

```python
import cv2
import matplotlib.pyplot as plt


# Path to the single evaluation image
evaluation_image = '/content/drive/MyDrive/AER850/data/evaluation/rasppi.jpg'
# Path to save the output image with the new name
output_image_path = '/content/drive/MyDrive/AER850/data/evaluation/epoch_185_imgsize_960_1.jpg'

# Run prediction and save results
results = model.predict(evaluation_image, save=True)

# Display the annotated image with bounding boxes, component names, and confidence scores
img = cv2.imread(results[0].path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Define lighter colors that stand out against dark green background
colors = [(255, 182, 193), (216, 191, 216), (173, 216, 230), (240, 128, 128)]  # Light pink, light purple, light blue, light coral

# Access bounding boxes, confidence scores, and class labels
for i, (*xyxy, conf, cls) in enumerate(results[0].boxes.data):
    x1, y1, x2, y2 = map(int, xyxy)  # Bounding box coordinates
    label = model.names[int(cls)]    # Class label
    color = colors[i % len(colors)]  # Cycle through colors

    # Draw bounding box and add component name and confidence score to the image
    cv2.rectangle(img, (x1, y1), (x2, y2), color, 4)  # Increase thickness from 2 to 4
    cv2.putText(img, f'{label} {conf:.2f}', (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

# Save the annotated image to Google Drive with the new name
cv2.imwrite(output_image_path, cv2.cvtColor(img, cv2.COLOR_RGB2BGR))

plt.imshow(img)
plt.axis('off')
plt.show()
```
