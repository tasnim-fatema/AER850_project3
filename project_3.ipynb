```python
# Step 1: Install required libraries
!pip install ultralytics

# Step 2: Import necessary packages
from ultralytics import YOLO
from google.colab import drive
import os

# Step 3: Mount Google Drive to access your dataset
drive.mount('/content/drive')
```

```python
# Step 4: Configure and load the YOLOv8 nano model
# Initialize the model with the YOLOv8-nano configuration
model = YOLO('yolov8n.pt')  # Ensure this model file exists or use a pre-trained version

# Step 5: Define the dataset location
DATASET_PATH = '/content/drive/MyDrive/AER850/data/data.yaml'  # Path to your dataset configuration file

# Step 6: Configure hyperparameters for training
EPOCHS = 120
BATCH_SIZE = 8
IMG_SIZE = 960
MODEL_NAME = "pcb_detection_yolov8n"  # Name for the trained model


```

```python
# Step 7: Train the model
model.train(
    data=DATASET_PATH,   # Path to the dataset configuration file
    epochs=EPOCHS,       # Number of training epochs
    batch=BATCH_SIZE,    # Batch size for training
    imgsz=IMG_SIZE,      # Input image size
    name=MODEL_NAME      # Model name for saving results
)
```

```python
# Step 8: Evaluate the model
metrics = model.val()
```

```python
import cv2
import matplotlib.pyplot as plt
import random
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Path to the single evaluation image
evaluation_image = '/content/drive/MyDrive/AER850/data/evaluation/rasppi.jpg'
# Path to save the output image
output_image_path = '/content/drive/MyDrive/AER850/data/evaluation/annotated_rasppi.jpg'

# Run prediction and save results
results = model.predict(evaluation_image, save=True)

# Display the annotated image with bounding boxes, component names, and confidence scores
img = cv2.imread(results[0].path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Assign colors to each unique label
label_colors = {model.names[int(cls)]: (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)) for *_, cls in results[0].boxes.data}

# Access bounding boxes, confidence scores, and class labels
for *xyxy, conf, cls in results[0].boxes.data:
    x1, y1, x2, y2 = map(int, xyxy)  # Bounding box coordinates
    label = model.names[int(cls)]    # Class label
    color = label_colors[label]      # Component-specific color

    # Draw bounding box and add component name and confidence score to the image
    cv2.rectangle(img, (x1, y1), (x2, y2), color, 3)
    cv2.putText(img, f'{label} {conf:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.5, color, 3)

# Save the annotated image to Google Drive
cv2.imwrite(output_image_path, cv2.cvtColor(img, cv2.COLOR_RGB2BGR))

plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import random
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Path to the single evaluation image
evaluation_image = '/content/drive/MyDrive/AER850/data/evaluation/arduno.jpg'
# Path to save the output image
output_image_path = '/content/drive/MyDrive/AER850/data/evaluation/annotated_arduno.jpg'

# Run prediction and save results
results = model.predict(evaluation_image, save=True)

# Display the annotated image with bounding boxes, component names, and confidence scores
img = cv2.imread(results[0].path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Define specific colors for components that stand out against the blue background
colors = [(255, 0, 0), (255, 255, 0), (0, 255, 0), (255, 0, 255)]
label_colors = {}

# Access bounding boxes, confidence scores, and class labels
for i, (*xyxy, conf, cls) in enumerate(results[0].boxes.data):
    x1, y1, x2, y2 = map(int, xyxy)  # Bounding box coordinates
    label = model.names[int(cls)]    # Class label

    # Assign color to component if not already assigned
    if label not in label_colors:
        label_colors[label] = colors[i % len(colors)]

    color = label_colors[label]  # Component-specific color

    # Draw bounding box and add component name and confidence score to the image
    cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
    cv2.putText(img, f'{label} {conf:.2f}', (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

# Save the annotated image to Google Drive
cv2.imwrite(output_image_path, cv2.cvtColor(img, cv2.COLOR_RGB2BGR))

plt.imshow(img)
plt.axis('off')
plt.show()
```
